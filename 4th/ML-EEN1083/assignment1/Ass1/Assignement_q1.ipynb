```
path = '/content/drive/MyDrive/Updated Datasets/Merged 2023.csv'
dataset23 = pd.read_csv(path)
path = '/content/drive/MyDrive/Updated Datasets/Merged 2024.csv'
dataset24 = pd.read_csv(path)
```

```
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'Date' column to datetime
dataset23['Date'] = pd.to_datetime(dataset23['Date'], format='%d/%m/%Y')  # Adjust format if necessary
dataset24['Date'] = pd.to_datetime(dataset24['Date'], format='%d/%m/%Y')  # Adjust format if necessary
# Extract the day of the week
dataset23['Day_of_Week'] = dataset23['Date'].dt.day_name()
dataset24['Day_of_Week'] = dataset24['Date'].dt.day_name()
# Calculate daily means for gas and electricity usage
# Summing hourly values for each day
gas_cols23 = [col for col in dataset23.columns if 'Gas' in col]
electricity_cols23 = [col for col in dataset23.columns if 'Electricity' in col]
gas_cols24 = [col for col in dataset24.columns if 'Gas' in col]
electricity_cols24 = [col for col in dataset24.columns if 'Electricity' in col]
dataset23['Daily_Gas'] = dataset23[gas_cols23].mean(axis=1)
dataset23['Daily_Electricity'] = dataset23[electricity_cols23].mean(axis=1)
dataset24['Daily_Gas'] = dataset24[gas_cols24].mean(axis=1)
dataset24['Daily_Electricity'] = dataset24[electricity_cols24].mean(axis=1)
# Melt the data for easier plotting with Seaborn
melted_data23 = pd.melt(dataset23, id_vars=['Day_of_Week'],
                      value_vars=['Daily_Gas', 'Daily_Electricity'],
                      var_name='Energy_Type', value_name='Usage')
melted_data24 = pd.melt(dataset24, id_vars=['Day_of_Week'],
                      value_vars=['Daily_Gas', 'Daily_Electricity'],
                      var_name='Energy_Type', value_name='Usage')
# Create a boxplot for daily usage by day of the week
plt.figure(figsize=(12, 6))
sns.boxplot(x='Day_of_Week', y='Usage', hue='Energy_Type', data=melted_data23,
            order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Daily Energy Usage by Day of the Week')
plt.ylabel('Energy Usage')
plt.xlabel('Day of the Week')
plt.legend(title='Energy Type')
plt.show()
```

```
from scipy.stats import pearsonr, spearmanr

# Create a scatter plot with density contours
plt.figure(figsize=(10, 6))
sns.kdeplot(
    x=dataset23['Daily_Gas'],
    y=dataset23['Daily_Electricity'],
    cmap="Blues", fill=True, alpha=0.5
)
sns.scatterplot(
    x=dataset23['Daily_Gas'],
   y=dataset23['Daily_Electricity'],
   color="red", alpha=0.6, edgecolor=None
)
plt.title('Relationship Between Gas and Electricity Usage')
plt.xlabel('Daily Mean Gas Usage')
plt.ylabel('Daily Mean Electricity Usage')
plt.show()

# Calculate correlation coefficients
pearson_corr, _ = pearsonr(dataset23['Daily_Gas'], dataset23['Daily_Electricity'])
spearman_corr, _ = spearmanr(dataset23['Daily_Gas'], dataset23['Daily_Electricity'])

# Print correlation results
print(f"Pearson Correlation Coefficient: {pearson_corr:.2f}")
print(f"Spearman Correlation Coefficient: {spearman_corr:.2f}")
```
